```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
```

```python
csv_file_path_3 = '/content/drive/MyDrive/Major_Project/Utility_Matrices/Utility_Matrix_5000_0.8.csv'
df_user = pd.read_csv(csv_file_path_3,index_col=0)
df_user.head()
```

```python
df_user = df_user.fillna(0)
df_user
```

```python
num_clusters = 2
kmeans = KMeans(n_clusters=num_clusters)
df_user_np = df_user
kmeans.fit(df_user_np)

# Assigning cluster labels to items
item_clusters = kmeans.labels_
print(item_clusters)

# Finding the centroid of each cluster
cluster_centers = kmeans.cluster_centers_

cluster_centers
```

```python
def euclidean_distance(point1, point2):
    point1 = np.array(point1)
    point2 = np.array(point2)
    squared_diff = (point1 - point2) ** 2
    distance = np.sqrt(np.sum(squared_diff))
    return distance

def recommend_user(user_ratings):
    distances = [euclidean_distance(user_ratings, point) for point in cluster_centers]
    print(distances)

    # Find the cluster with the minimum distance
    cluster_index = np.argmin(distances)
    print(cluster_index)


    # Items in the closest cluster are potential recommendations
    recommended_items = np.where(item_clusters == cluster_index)[0]

    return recommended_items

print(recommend_user(np.array([1,0,3,4,1])))

```
