```python
num_cols = len(df_city['City'].unique())
num_cols
```

```python
def generate_synthetic_utility_matrix(users, locations, density):
    matrix = np.zeros((users, locations))

    for i in range(users):
        for j in range(locations):
          n = np.random.rand()
          if n <= density:
            matrix[i, j] = np.random.randint(1, 6)

    return matrix
```

```python
def generate_csv_file(num_users,sparsity_density,file_path):
  num_locations = num_cols
  utility_matrix = generate_synthetic_utility_matrix(num_users, num_locations, sparsity_density)
  um_pd = pd.DataFrame(utility_matrix)
  um_pd.columns = df_city['City']
  um_pd.index = ['user_' + str(idx) for idx in um_pd.index]
  um_pd = um_pd.rename_axis(index='Users', columns='Locations')
  um_pd.to_csv(file_path)

```

```python

num_users = 100000
sparsity_density = 0.8
generate_csv_file(num_users,sparsity_density,'/content/drive/MyDrive/Major_Project/Utility_Matrices/Utility_Matrix_'+str(num_users)+'_'+str(sparsity_density)+'.csv')
```
